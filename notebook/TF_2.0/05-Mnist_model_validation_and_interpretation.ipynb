## Model interpretation
### Helper functions

```python
def evaluation_indices(y_pred, y_test):
    # function for getting correctly and incorrectly classified indices
    index = 0
    correctly_classified_indices = []
    misclassified_indices = []
    for label, predict in zip(y_test, y_pred):
        if label != predict: 
            misclassified_indices.append(index)
        else:
            correctly_classified_indices.append(index)
        index +=1
    return (correctly_classified_indices, misclassified_indices)
```

### Select examples of correct and incorrect classified classes

```python
correctly_classified_indices, misclassified_indices = evaluation_indices(predict_classes, true_classes)
```

### Use LIME for correct classified classes

```python
# creating issue with 2.0
#logging.set_verbosity(logging.INFO)
logging.debug('test')
logging.info('test')
```

```python
def new_predict_fn(images):
    # create dummy label
    labels=np.empty([len(images)], dtype=int)
    
    # RGB ( None, 28, 28, 3) to Grey images (None, 28, 28)
    images=rgb2gray(images)
    
    # Applied preprocessing
    dataset = mnist_v1.input_mnist_array_dataset_fn(images, 
                                                    labels, 
                                                    FLAGS,mode=tf.estimator.ModeKeys.EVAL, 
                                                    batch_size=len(labels))
    # debug
    debug_info=False

    for features, labels in dataset:
        if debug_info:
            start_time = time.clock()
            print('iteration n=', n, 'execution time:', time.clock() - start_time, 'seconds')
            print(features['dense_input'].shape)
            print(labels.shape)
            print('first label of the batch',np.argmax(labels[0]),'\n')
                
        # we just need the full dataset so one epoch
        break
                
    return best_model.predict(features)
```

```python
plt.figure(figsize=(25,5))
#shuffle(correctly_classified_indices)
for plot_index, good_index in enumerate(correctly_classified_indices[0:5]):
    plt.subplot(1, 5, plot_index + 1)
    plt.imshow(x_test[good_index], cmap=plt.cm.gray)
    plt.title('Predicted: {}, Actual: {}'.format(labels_index[predict_classes[good_index]], 
                                                 labels_index[true_classes[good_index]]), fontsize = 15)
```

```python
plt.figure(figsize=(25,5))
#shuffle(correctly_classified_indices)
for plot_index, good_index in enumerate(correctly_classified_indices[0:5]):
    plt.subplot(1, 5, plot_index + 1)
    explainer = lime_image.LimeImageExplainer()
    explanation = explainer.explain_instance(gray2rgb(x_test[good_index]), new_predict_fn, top_labels=10, hide_color=0, num_samples=1000)
    temp, mask = explanation.get_image_and_mask(0, positive_only=True, num_features=10, hide_rest=False)
    x = mark_boundaries(temp / 2 + 0.5, mask)
    plt.imshow((x * 255).astype(np.uint8), interpolation='none')
    plt.title('Predicted: {}, Actual: {}'.format(labels_index[predict_classes[good_index]], 
                                                 labels_index[true_classes[good_index]]), fontsize = 15)
```
